# 100 Questões Teóricas Comuns em Testes Técnicos
## JAVA - Português

Este documento apresenta 100 questões teóricas comuns que podem surgir em testes técnicos de Java. As questões abrangem uma variedade de tópicos, incluindo a JVM, coleções, tratamento de exceções, programação concorrente, frameworks populares como Spring e Hibernate, e muito mais. Cada pergunta é acompanhada de respostas detalhadas para auxiliar no estudo e na revisão.

## 100 Questões de Java

1. **O que é a JVM e qual é sua função principal?**  
   **Resposta:** A JVM (Java Virtual Machine) é uma máquina virtual que executa programas Java. Sua função principal é traduzir o bytecode Java em código de máquina específico da plataforma para que possa ser executado pelo sistema operacional.

2. **O que é o JDK e o JRE?**  
   **Resposta:** O JDK (Java Development Kit) é um pacote de software necessário para desenvolver aplicativos em Java, incluindo compilador, bibliotecas e ferramentas de desenvolvimento. O JRE (Java Runtime Environment) é um conjunto de ferramentas para execução de aplicativos Java, incluindo a JVM, bibliotecas de classes e outros componentes de suporte.

3. **Explique a diferença entre sobrecarga (overloading) e sobrescrita (overriding) de métodos.**  
   **Resposta:** Sobrecarga de métodos ocorre quando dois ou mais métodos em uma classe têm o mesmo nome, mas parâmetros diferentes. Sobrescrita de métodos acontece quando uma subclasse fornece uma implementação específica de um método já definido em sua superclasse.

4. **O que são exceções em Java e como são tratadas?**  
   **Resposta:** Exceções são eventos anormais que ocorrem durante a execução de um programa e interrompem o fluxo normal do programa. Em Java, elas são tratadas usando blocos try-catch, onde o código propenso a exceções é colocado dentro do bloco try e o tratamento da exceção é feito no bloco catch.

5. **O que é a herança e como é implementada em Java?**  
   **Resposta:** Herança é um princípio da programação orientada a objetos onde uma classe (subclasse) pode herdar campos e métodos de outra classe (superclasse). Em Java, é implementada usando a palavra-chave extends.

6. **O que é uma interface e como ela difere de uma classe abstrata?**  
   **Resposta:** Uma interface é um contrato que especifica métodos que uma classe deve implementar, mas não contém implementação dos métodos. Uma classe abstrata pode ter métodos concretos e abstratos e pode manter estado, enquanto uma interface não pode.

7. **Explique o conceito de polimorfismo.**  
   **Resposta:** Polimorfismo é a capacidade de um objeto assumir várias formas. Em Java, é manifestado através da sobrescrita de métodos, onde uma superclasse define um método que pode ser implementado de maneiras diferentes por subclasses.

8. **O que é encapsulamento e como é alcançado em Java?**  
   **Resposta:** Encapsulamento é o princípio de ocultar os detalhes de implementação interna de um objeto e expor apenas os métodos necessários. Em Java, é alcançado usando modificadores de acesso como private, protected e public.

9. **O que são coleções em Java?**  
   **Resposta:** Coleções são estruturas de dados que armazenam grupos de objetos. O framework de coleções em Java fornece várias interfaces e classes para manipular conjuntos de dados, como listas, conjuntos e mapas.

10. **Explique a diferença entre ArrayList e LinkedList.**  
    **Resposta:** ArrayList é baseado em arrays dinâmicos, oferecendo acesso rápido a elementos, mas inserções e remoções são lentas devido ao deslocamento de elementos. LinkedList é baseado em uma lista duplamente ligada, oferecendo inserções e remoções rápidas, mas acesso lento a elementos devido à navegação sequencial.

11. **O que é o garbage collector em Java?**  
    **Resposta:** O garbage collector (GC) é um mecanismo automático de gerenciamento de memória que remove objetos não mais referenciados para liberar espaço na memória heap, evitando vazamento de memória.

12. **Explique o conceito de String imutável.**  
    **Resposta:** Em Java, a classe String é imutável, o que significa que uma vez criada, seu valor não pode ser alterado. Qualquer operação que modifique uma String cria uma nova instância de String.

13. **O que é o conceito de final em Java?**  
    **Resposta:** A palavra-chave final em Java pode ser usada com variáveis, métodos e classes. Uma variável final não pode ser reatribuída, um método final não pode ser sobrescrito, e uma classe final não pode ser estendida.

14. **Explique o que são lambdas e streams introduzidos no Java 8.**  
    **Resposta:** Lambdas são expressões que permitem tratar funções como objetos de primeira classe, facilitando a programação funcional. Streams são uma nova abstração para processar sequências de dados de forma declarativa, permitindo operações como filtro, mapeamento e redução.

15. **O que é um singleton e como é implementado em Java?**  
    **Resposta:** Singleton é um padrão de design que garante que uma classe tenha apenas uma instância e fornece um ponto global de acesso a essa instância. Em Java, é implementado criando um construtor privado e um método estático para retornar a instância única.

16. **Explique a diferença entre == e equals() em Java.**  
    **Resposta:** == compara referências de objetos para verificar se apontam para o mesmo objeto na memória. O método equals() compara o conteúdo dos objetos para verificar se são logicamente equivalentes.

17. **O que são generics em Java?**  
    **Resposta:** Generics são uma maneira de definir classes, interfaces e métodos com tipos de parâmetros, permitindo que o código seja reutilizado com diferentes tipos de dados, aumentando a segurança e reduzindo a necessidade de conversões explícitas.

18. **Explique a diferença entre HashMap e TreeMap.**  
    **Resposta:** HashMap armazena pares chave-valor sem ordenação garantida, usando um hash para determinar a posição dos elementos. TreeMap armazena pares chave-valor em uma estrutura de árvore, garantindo a ordenação natural das chaves.

19. **O que é a serialização em Java?**  
    **Resposta:** Serialização é o processo de converter um objeto em uma sequência de bytes para que ele possa ser salvo em um arquivo ou transmitido pela rede. A desserialização é o processo inverso, reconstruindo o objeto a partir da sequência de bytes.

20. **O que é o método hashCode() e por que ele é importante?**  
    **Resposta:** O método hashCode() retorna um valor hash inteiro que é usado para a busca eficiente em estruturas de dados baseadas em hash, como HashMap. É importante que hashCode() seja consistente com equals() para manter a integridade dos dados.

21. **Explique o conceito de concorrência e como é gerenciado em Java.**  
    **Resposta:** Concorrência é a execução de várias tarefas simultaneamente. Em Java, é gerenciado através de threads, que podem ser criadas usando a classe Thread ou implementando a interface Runnable. Java também fornece o pacote java.util.concurrent para facilitar a programação concorrente.

22. **O que é um deadlock e como ele pode ser evitado?**  
    **Resposta:** Deadlock ocorre quando duas ou mais threads estão bloqueadas indefinidamente, esperando por recursos que estão sendo mantidos por outras threads. Pode ser evitado usando técnicas como evitar a espera circular, usando temporizadores para operações de bloqueio ou implementando políticas de ordenação de recursos.

23. **Explique a diferença entre synchronized method e synchronized block.**  
    **Resposta:** Um método synchronized garante que apenas uma thread pode executá-lo por vez, bloqueando o objeto inteiro. Um bloco synchronized permite sincronizar uma seção específica do código, oferecendo maior controle e desempenho ao permitir que outras partes do método sejam executadas simultaneamente.

24. **O que é o Spring Framework?**  
    **Resposta:** Spring Framework é um framework Java de código aberto que fornece infraestrutura abrangente para o desenvolvimento de aplicativos empresariais, promovendo boas práticas como injeção de dependência, programação orientada a aspectos e suporte para transações declarativas.

25. **Explique o conceito de injeção de dependência (DI) no Spring.**  
    **Resposta:** Injeção de dependência é um padrão de design onde um objeto recebe suas dependências de um contêiner de injeção de dependência em vez de criá-las diretamente. No Spring, isso é realizado através de anotações como @Autowired e configurações XML.

26. **O que são Beans no contexto do Spring Framework?**  
    **Resposta:** Beans são objetos gerenciados pelo contêiner do Spring. Eles são configurados, instanciados e gerenciados pelo Spring Framework, permitindo injeção de dependência e ciclo de vida controlado.

27. **Explique o que é o Hibernate e como ele se relaciona com o JPA.**  
    **Resposta:** Hibernate é um framework de mapeamento objeto-relacional (ORM) que facilita a persistência de objetos Java em bancos de dados relacionais. JPA (Java Persistence API) é uma especificação que define uma API padrão para ORM. Hibernate é uma implementação popular da JPA.

28. **O que é a anotação @Entity em JPA?**  
    **Resposta:** A anotação @Entity é usada para especificar que uma classe Java representa uma entidade mapeada para uma tabela em um banco de dados relacional.

29. **Explique a diferença entre fetchType.EAGER e fetchType.LAZY em JPA.**  
    **Resposta:** fetchType.EAGER significa que as associações são carregadas imediatamente junto com a entidade principal, enquanto fetchType.LAZY carrega as associações sob demanda, quando são acessadas pela primeira vez.

30. **O que é o Maven e qual é sua função no desenvolvimento Java?**  
    **Resposta:** Maven é uma ferramenta de gerenciamento de projetos e automação de builds para Java. Ele gerencia dependências de bibliotecas, compila o código, executa testes e gera pacotes de distribuição.

31. **Explique a estrutura de um arquivo pom.xml no Maven.**  
    **Resposta:** O pom.xml (Project Object Model) é o arquivo de configuração central do Maven. Ele define as dependências, plugins, propriedades do projeto e informações do build, como a versão do projeto e os repositórios.

32. **O que é o conceito de microsserviços?**  
    **Resposta:** Microsserviços são uma arquitetura de software onde uma aplicação é dividida em pequenos serviços independentes que se comunicam através de APIs. Cada serviço é responsável por uma funcionalidade específica e pode ser desenvolvido, implantado e escalado de forma independente.

33. **Explique a diferença entre REST e SOAP.**  
    **Resposta:** REST (Representational State Transfer) é um estilo de arquitetura que usa HTTP para comunicação entre serviços, focando em simplicidade e uso de operações CRUD. SOAP (Simple Object Access Protocol) é um protocolo baseado em XML para troca de informações estruturadas, com suporte para segurança e transações.

34. **O que é uma API e qual é sua importância?**  
    **Resposta:** API (Application Programming Interface) é um conjunto de definições e protocolos que permitem que diferentes softwares se comuniquem. É importante para a integração de sistemas, permitindo que funcionalidades de um aplicativo sejam acessadas por outros.

35. **O que é o Spring Boot e quais são seus benefícios?**  
    **Resposta:** Spring Boot é uma extensão do Spring Framework que facilita a criação de aplicativos standalone, produção-ready, com configuração mínima. Benefícios incluem inicialização rápida, configuração automática e um conjunto integrado de ferramentas de desenvolvimento.

36. **Explique o que é uma anotação @SpringBootApplication.**  
    **Resposta:** @SpringBootApplication é uma anotação de conveniência que combina @Configuration, @EnableAutoConfiguration e @ComponentScan. Ela configura automaticamente a aplicação Spring Boot.

37. **O que é o conceito de profiles no Spring Boot?**  
    **Resposta:** Profiles no Spring Boot permitem que diferentes configurações sejam aplicadas em diferentes ambientes (desenvolvimento, teste, produção). Eles são definidos usando a anotação @Profile ou no arquivo de propriedades application.properties.

38. **Explique a diferença entre @Component, @Service e @Repository no Spring.**  
    **Resposta:** Todas são anotações de estereótipos usadas para marcar classes como beans do Spring. @Component é uma anotação genérica, @Service indica que a classe fornece serviços de negócios e @Repository indica que a classe interage com o banco de dados, incluindo tradução de exceções.

39. **O que é um RestController no Spring Boot?**  
    **Resposta:** @RestController é uma anotação que combina @Controller e @ResponseBody, indicando que a classe é um controlador que trata requisições HTTP e retorna dados diretamente, geralmente em formato JSON ou XML.

40. **Explique a diferença entre GET, POST, PUT e DELETE em uma API REST.**  
    **Resposta:** GET é usado para recuperar recursos, POST para criar novos recursos, PUT para atualizar recursos existentes e DELETE para remover recursos. Esses métodos correspondem às operações CRUD.

41. **O que é uma transação e como é gerenciada no Spring?**  
    **Resposta:** Uma transação é uma sequência de operações de banco de dados que são tratadas como uma unidade única, garantindo consistência e integridade. No Spring, transações são gerenciadas usando a anotação @Transactional para demarcar métodos transacionais.

42. **O que é o JUnit e qual é sua função?**  
    **Resposta:** JUnit é um framework de teste unitário para Java. Ele permite criar e executar testes automatizados para garantir que o código funcione conforme esperado.

43. **Explique o conceito de mocking em testes.**  
    **Resposta:** Mocking é uma técnica de teste onde objetos simulados (mocks) são criados para imitar o comportamento de objetos reais, permitindo testar componentes de forma isolada. Ferramentas como Mockito são usadas para criar mocks.

44. **O que é um arquivo WAR e um arquivo JAR?**  
    **Resposta:** Um arquivo JAR (Java Archive) é um pacote que contém classes, bibliotecas e recursos de um aplicativo Java. Um arquivo WAR (Web Application Archive) é um pacote específico para aplicativos web, contendo classes, bibliotecas, recursos e configurações de implantação.

45. **Explique o conceito de injeção de dependência (DI) e inversão de controle (IoC).**  
    **Resposta:** Injeção de dependência é um padrão onde um objeto recebe suas dependências de um contêiner IoC em vez de criá-las diretamente. Inversão de controle é um princípio onde o controle do fluxo do programa é invertido, sendo gerenciado por um contêiner ou framework.

46. **O que é o Maven e qual é sua função no desenvolvimento Java?**  
    **Resposta:** Maven é uma ferramenta de gerenciamento de projetos e automação de builds para Java. Ele gerencia dependências de bibliotecas, compila o código, executa testes e gera pacotes de distribuição.

47. **O que é o Gradle e como ele se compara ao Maven?**  
    **Resposta:** Gradle é uma ferramenta de automação de builds que usa uma linguagem de script baseada em Groovy ou Kotlin. Comparado ao Maven, Gradle oferece mais flexibilidade e performance, suportando builds incrementais e paralelos.

48. **Explique o conceito de serialização e desserialização.**  
    **Resposta:** Serialização é o processo de converter um objeto em uma sequência de bytes para que ele possa ser salvo em um arquivo ou transmitido pela rede. Desserialização é o processo inverso, reconstruindo o objeto a partir da sequência de bytes.

49. **O que é o Java Memory Model (JMM)?**  
    **Resposta:** O Java Memory Model define como threads interagem através da memória e garante a visibilidade das mudanças feitas por uma thread para outras threads. Ele fornece regras sobre como os dados são armazenados e recuperados na memória.

50. **O que é o volatile em Java?**  
    **Resposta:** volatile é uma palavra-chave usada para marcar uma variável que pode ser acessada por várias threads. Ela garante que as leituras e escritas na variável sejam feitas diretamente na memória principal, assegurando visibilidade das mudanças entre threads.

51. **Explique o conceito de imutabilidade em Java.**  
    **Resposta:** Imutabilidade é uma propriedade onde o estado de um objeto não pode ser alterado após sua criação. Objetos imutáveis são thread-safe por natureza, pois não podem ser modificados por múltiplas threads.

52. **O que é a palavra-chave this em Java?**  
    **Resposta:** this é uma referência ao objeto atual dentro de um método ou construtor. É usado para diferenciar entre variáveis de instância e parâmetros de método ou para invocar outros construtores na mesma classe.

53. **O que é a palavra-chave super em Java?**  
    **Resposta:** super é uma referência à superclasse imediata de um objeto. É usado para acessar membros da superclasse que foram sobrescritos pela subclasse ou para invocar o construtor da superclasse.

54. **Explique o conceito de classe aninhada e classe interna.**  
    **Resposta:** Uma classe aninhada é uma classe definida dentro de outra classe. Se não for estática, é chamada de classe interna e pode acessar membros da classe externa. Classes aninhadas podem ser usadas para agrupar classes logicamente relacionadas e aumentar a encapsulação.

55. **O que é uma enumeração em Java e como é usada?**  
    **Resposta:** Uma enumeração (enum) é um tipo especial de classe que representa um conjunto fixo de constantes. É usada para definir coleções de valores constantes, melhorando a legibilidade e segurança do código.

56. **Explique a diferença entre um Set e uma List em Java.**  
    **Resposta:** Set é uma coleção que não permite elementos duplicados e não garante ordem específica dos elementos. List é uma coleção ordenada que permite elementos duplicados e acessa elementos por índice.

57. **O que é um Map em Java?**  
    **Resposta:** Map é uma coleção que mapeia chaves únicas a valores. Ele não permite chaves duplicadas e cada chave pode mapear para apenas um valor. Exemplos incluem HashMap e TreeMap.

58. **Explique a diferença entre ConcurrentHashMap e HashMap.**  
    **Resposta:** HashMap não é thread-safe e deve ser sincronizado manualmente para uso concorrente. ConcurrentHashMap é uma implementação thread-safe que permite acesso concorrente eficiente sem bloqueios em toda a estrutura.

59. **O que é a anotação @FunctionalInterface em Java?**  
    **Resposta:** @FunctionalInterface é usada para marcar uma interface que contém exatamente um método abstrato. Essa anotação ajuda a garantir que a interface seja adequada para uso em expressões lambda e referências de método.

60. **Explique o conceito de expressão lambda.**  
    **Resposta:** Uma expressão lambda é uma forma concisa de representar uma função anônima, permitindo passar comportamento como argumento para métodos. Elas foram introduzidas no Java 8 para facilitar a programação funcional.

61. **O que são referências de método em Java?**  
    **Resposta:** Referências de método são uma forma de simplificar expressões lambda, referenciando diretamente métodos existentes por seu nome. Elas são usadas com a sintaxe `ClassName::methodName`.

62. **Explique o que são Optional e suas vantagens.**  
    **Resposta:** Optional é uma classe contêiner usada para representar a presença ou ausência de um valor. Ela ajuda a evitar `NullPointerException` e torna o código mais expressivo ao indicar claramente quando um valor pode ser ausente.

63. **O que é um Stream em Java e como ele é usado?**  
    **Resposta:** Stream é uma abstração para processar sequências de dados de forma declarativa, usando operações como filtro, mapeamento e redução. Eles permitem realizar operações complexas de forma concisa e eficiente.

64. **Explique a diferença entre `parallelStream()` e `stream()`.**  
    **Resposta:** `stream()` cria um fluxo sequencial que processa elementos de forma síncrona. `parallelStream()` cria um fluxo paralelo que divide a tarefa em múltiplos threads para execução simultânea, aproveitando melhor os processadores multicore.

65. **O que é o conceito de default methods em interfaces?**  
    **Resposta:** Métodos default em interfaces permitem fornecer implementações padrão para métodos diretamente na interface. Isso permite adicionar novos métodos às interfaces sem quebrar implementações existentes.

66. **Explique o que são annotations e como são usadas em Java.**  
    **Resposta:** Annotations são metadados que fornecem informações adicionais sobre o código. Elas são usadas para instruir o compilador, ferramentas de geração de código e frameworks sobre como tratar classes, métodos e campos.

67. **O que é o conceito de try-with-resources?**  
    **Resposta:** `try-with-resources` é uma declaração `try` que gerencia automaticamente recursos como `InputStream` e `OutputStream`. Qualquer recurso que implemente a interface `AutoCloseable` pode ser usado, garantindo fechamento automático ao final do bloco.

68. **Explique o que é a reflection em Java.**  
    **Resposta:** Reflection é a capacidade de inspecionar e manipular a estrutura de classes, métodos e campos em tempo de execução. É usado para criar frameworks, ferramentas de teste e bibliotecas que exigem introspecção dinâmica do código.

69. **O que é o Spring Data JPA?**  
    **Resposta:** Spring Data JPA é um módulo do Spring que simplifica o acesso a dados em aplicativos Java, fornecendo uma camada de abstração sobre o JPA. Ele permite criar repositórios de dados com consultas poderosas sem escrever SQL manualmente.

70. **Explique a diferença entre `findById` e `getOne` no Spring Data JPA.**  
    **Resposta:** `findById` retorna um `Optional` contendo a entidade correspondente ou vazio se não encontrada. `getOne` retorna uma entidade proxy que é carregada sob demanda e pode lançar `EntityNotFoundException` se a entidade não existir ao acessar seus campos.

71. **O que é o Quarkus e quais são seus benefícios?**  
    **Resposta:** Quarkus é um framework Java otimizado para a execução em ambientes de contêineres e nuvem, oferecendo tempos de inicialização rápidos e baixa utilização de memória. Ele suporta padrões e bibliotecas Java existentes, integrando bem com tecnologias como GraalVM para compilação nativa.

72. **Explique o conceito de arquétipo em Maven.**  
    **Resposta:** Um arquétipo em Maven é um modelo de projeto que fornece uma estrutura básica e configurações iniciais para criar novos projetos. Ele facilita a criação de projetos consistentes e padrão.

73. **O que é a anotação @Bean no Spring?**  
    **Resposta:** A anotação @Bean é usada para declarar um método que retorna um bean gerenciado pelo contêiner do Spring. É comumente usada em classes de configuração para definir beans manualmente.

74. **Explique o conceito de EntityManager em JPA.**  
    **Resposta:** EntityManager é a interface principal para interagir com o contexto de persistência em JPA. Ele fornece métodos para realizar operações CRUD, gerenciar transações e consultar entidades.

75. **O que é a anotação @RestController e como ela difere de @Controller?**  
    **Resposta:** @RestController é uma combinação de @Controller e @ResponseBody, usada para criar controladores que retornam dados diretamente (geralmente JSON ou XML) em vez de renderizar uma visão. @Controller é usada para controladores tradicionais que retornam views.

76. **O que é a anotação @RequestMapping no Spring MVC?**  
    **Resposta:** @RequestMapping é usada para mapear requisições HTTP a métodos manipuladores em controladores Spring MVC. Pode ser aplicada a classes ou métodos e suporta configuração de caminhos, métodos HTTP, parâmetros e cabeçalhos.

77. **Explique o conceito de Aspect-Oriented Programming (AOP) no Spring.**  
    **Resposta:** AOP é uma abordagem de programação que permite separar preocupações transversais (como logging, segurança e transações) do código principal. No Spring, é implementado usando aspectos e interceptores para aplicar comportamento adicional a métodos existentes.

78. **O que é a anotação @Transactional no Spring?**  
    **Resposta:** A anotação @Transactional é usada para demarcar métodos ou classes como transacionais. Ela permite definir o escopo das transações e configurar atributos como isolamento, propagação e tempo limite.

79. **Explique o conceito de Circuit Breaker e como ele é implementado no Spring Cloud.**  
    **Resposta:** Circuit Breaker é um padrão que ajuda a prevenir falhas em cascata em sistemas distribuídos, interrompendo chamadas para serviços com falhas persistentes. No Spring Cloud, é implementado usando bibliotecas como Netflix Hystrix ou Resilience4j.

80. **O que é o Spring Security e quais são seus principais componentes?**  
    **Resposta:** Spring Security é um framework de segurança para aplicativos Java, fornecendo autenticação, autorização e proteção contra ataques comuns. Seus principais componentes incluem filtros de segurança, provedores de autenticação e configuração de segurança baseada em anotações.

81. **O que é o OAuth2 e como ele é usado no Spring Security?**  
    **Resposta:** OAuth2 é um protocolo de autorização que permite acesso seguro a recursos protegidos em nome de um usuário. No Spring Security, é usado para implementar autenticação e autorização delegada, permitindo que aplicativos acessem APIs de terceiros de forma segura.

82. **Explique o conceito de Feign Client no Spring Cloud.**  
    **Resposta:** Feign Client é uma biblioteca declarativa de clientes HTTP, integrada ao Spring Cloud, que facilita a chamada de serviços RESTful. Ele permite definir interfaces Java com anotações que são automaticamente traduzidas em chamadas HTTP.

83. **O que é a anotação @Configuration no Spring?**  
    **Resposta:** A anotação @Configuration é usada para marcar classes como fontes de definições de beans no Spring. Essas classes podem definir métodos anotados com @Bean que criam e configuram beans gerenciados pelo contêiner do Spring.

84. **Explique a diferença entre @Primary e @Qualifier no Spring.**  
    **Resposta:** @Primary é usado para marcar um bean como a escolha padrão quando há múltiplos candidatos para injeção. @Qualifier é usado para desambiguar beans específicos quando múltiplos candidatos existem, especificando qual bean deve ser injetado.

85. **O que é o Spring Batch e para que é usado?**  
    **Resposta:** Spring Batch é um framework de processamento de lotes usado para desenvolver aplicativos robustos e reutilizáveis de processamento de dados em lote. Ele oferece suporte para operações de leitura, processamento e gravação de grandes volumes de dados de forma eficiente.

86. **Explique o conceito de hystrix no contexto do Spring Cloud.**  
    **Resposta:** Hystrix é uma biblioteca do Netflix para implementação de Circuit Breakers, que ajuda a gerenciar falhas e latências em sistemas distribuídos. No contexto do Spring Cloud, Hystrix é usado para fornecer resiliência e tolerância a falhas em serviços de microsserviços.

87. **O que é o Spring Boot Actuator?**  
    **Resposta:** Spring Boot Actuator é um conjunto de ferramentas que fornece monitoramento e gerenciamento de aplicativos Spring Boot em tempo de execução. Ele oferece endpoints para métricas, saúde do aplicativo, rastreamento de requisições e outras informações operacionais.

88. **Explique o conceito de Reactive Programming no Spring WebFlux.**  
    **Resposta:** Reactive Programming é um paradigma de programação assíncrona e orientada a eventos, focada em manipular fluxos de dados. No Spring WebFlux, é implementado usando o projeto Reactor, permitindo a criação de aplicativos web reativos e não-bloqueantes.

89. **O que é o Spring Cloud Config?**  
    **Resposta:** Spring Cloud Config é um serviço que fornece configuração externa e centralizada para aplicativos distribuídos. Ele permite que aplicativos carreguem suas configurações de um repositório central, como Git, em tempo de execução.

90. **Explique o conceito de service registry e service discovery no Spring Cloud.**  
    **Resposta:** Service Registry é um repositório onde serviços registram suas instâncias para permitir descoberta. Service Discovery é o processo pelo qual um serviço encontra a localização de outros serviços dinamicamente. No Spring Cloud, isso é implementado usando bibliotecas como Eureka.

91. **O que é o Ribbon no contexto do Spring Cloud?**  
    **Resposta:** Ribbon é uma biblioteca de balanceamento de carga cliente-side usada no Spring Cloud. Ela permite distribuir requisições entre múltiplas instâncias de um serviço para melhorar a escalabilidade e resiliência.

92. **Explique o que é o Zuul no Spring Cloud.**  
    **Resposta:** Zuul é um gateway de API do Netflix usado no Spring Cloud para roteamento dinâmico, balanceamento de carga, autenticação e outras funcionalidades. Ele age como um ponto de entrada único para todas as requisições em uma arquitetura de microsserviços.

93. **O que é o Spring Cloud Stream?**  
    **Resposta:** Spring Cloud Stream é um framework para construção de aplicativos de processamento de dados em tempo real usando sistemas de mensagens como Kafka e RabbitMQ. Ele abstrai detalhes de integração e facilita a criação de pipelines de dados distribuídos.

94. **Explique o conceito de event-driven architecture no contexto do Spring.**  
    **Resposta:** Event-driven architecture é um padrão de arquitetura onde componentes se comunicam através de eventos, permitindo acoplamento fraco e escalabilidade. No Spring, é implementado usando tecnologias como Spring Events, Spring Cloud Stream e sistemas de mensagens.

95. **O que é o Spring Cloud Gateway?**  
    **Resposta:** Spring Cloud Gateway é um gateway de API baseado em Spring WebFlux que fornece roteamento, segurança e resiliência para microsserviços. Ele oferece uma alternativa moderna e reativa ao Zuul.

96. **Explique o conceito de ConfigurationProperties no Spring Boot.**  
    **Resposta:** `@ConfigurationProperties` é uma anotação usada para mapear propriedades de configuração externas para objetos Java. Ela facilita a leitura e validação de configurações definidas em arquivos `application.properties` ou `application.yml`.

97. **O que é o Spring Cloud Sleuth?**  
    **Resposta:** Spring Cloud Sleuth é uma biblioteca que adiciona rastreamento distribuído a aplicativos Spring. Ele anota logs com informações de rastreamento, permitindo correlacionar logs entre diferentes serviços e identificar o fluxo de requisições.

98. **Explique a diferença entre @ControllerAdvice e @RestControllerAdvice no Spring.**  
    **Resposta:** `@ControllerAdvice` é uma anotação usada para definir lógica global de tratamento de exceções e configuração de controladores. `@RestControllerAdvice` é uma especialização que combina `@ControllerAdvice` e `@ResponseBody`, aplicando tratamento a controladores RESTful.

99. **O que é o Spring Boot DevTools?**  
    **Resposta:** Spring Boot DevTools é um conjunto de ferramentas que facilita o desenvolvimento de aplicativos Spring Boot, fornecendo recursos como reinicialização automática, desativação de cache e suporte aprimorado para desenvolvimento local.

100. **Explique o conceito de Spring Cloud Bus.**  
    **Resposta:** Spring Cloud Bus é uma ferramenta que conecta instâncias de aplicativos através de um barramento de mensagens, permitindo a comunicação e a propagação de eventos de configuração e outros eventos entre múltiplos serviços.
